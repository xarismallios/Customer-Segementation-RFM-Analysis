```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import datetime as dt

data = pd.read_excel("Online Retail.xlsx")
data.head()
data.describe()
```

```python
data= data[pd.notnull(data['CustomerID'])]
data=data.drop_duplicates()
data = data[(data['Quantity']>0)]
data
```

```python
final=data[['CustomerID','InvoiceDate','InvoiceNo','Quantity','UnitPrice']]
```

```python
final['TotalPrice'] = final['Quantity'] * final['UnitPrice']
PRESENT = dt.datetime(2011,12,10)
final['InvoiceDate'] = pd.to_datetime(final['InvoiceDate'])
final
```

```python
rfm= final.groupby('CustomerID').agg({'InvoiceDate': lambda date: (PRESENT - date.max()).days,
                                        'InvoiceNo': lambda num: len(num),
                                        'TotalPrice': lambda price: price.sum()})
```

```python
rfm.columns=['monetary','frequency','recency']
rfm['recency'] = rfm['recency'].astype(int)
```

```python
rfm
```

```python
rfm['r_quartile'] = pd.qcut(rfm['recency'], 4, ['1','2','3','4'])
rfm['f_quartile'] = pd.qcut(rfm['frequency'], 4, ['4','3','2','1'])
rfm['m_quartile'] = pd.qcut(rfm['monetary'], 4, ['4','3','2','1'])

rfm['RFM_Score'] = rfm.r_quartile.astype(str)+ rfm.f_quartile.astype(str) + rfm.m_quartile.astype(str)
rfm.head()
```
